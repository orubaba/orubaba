```python
import matplotlib.pyplot as plt

# Function to read coordinates from a file and plot
def plot_coordinates_from_file(file_path, title):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    x = []
    y = []
    for line in lines:
        if not line.startswith('#'):
            coords = line.split()
            x.append(float(coords[0]))
            y.append(float(coords[1]))

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, marker='o', linestyle='-', color='blue')
    plt.xlabel('Intrinsic Reaction Coordinate (IRC)')
    plt.ylabel('Total Energy (Hartree)')
    plt.title(title)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# File paths for the two datasets
file_path_data_1 = "/content/drive/MyDrive/new_ligands_csvs_folder/DOYMUC_LFeOOH_CH4_rxt_scan_cont2_TS_stab_ts_irc_forward_opt_tot_ener.txt"  # Replace 'path_to_data_1.txt' with the actual file path
file_path_data_2 = '/content/drive/MyDrive/new_ligands_csvs_folder/DOYMUC_LFeOOH_CH4_rxt_scan_cont2_TS_stab_ts_irc_reverse_opt_tot_ener.txt'  # Replace 'path_to_data_2.txt' with the actual file path

# Plotting the first set of coordinates
plot_coordinates_from_file(file_path_data_1, "Total Energy along IRC - Dataset 1")

# Plotting the second set of coordinates
plot_coordinates_from_file(file_path_data_2, "Total Energy along IRC - Dataset 2")
```

```python
import matplotlib.pyplot as plt

# Function to mirror coordinates along X-axis
def mirror_coordinates(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    x = []
    y = []
    for line in lines:
        if not line.startswith('#'):
            coords = line.split()
            x.append(float(coords[0]))
            y.append(float(coords[1]))

    # Create a mirrored version of the coordinates along the X-axis
    x_mirror = [-val for val in x]  # Mirror the x-coordinates

    return x, y, x_mirror

# File path for the dataset
file_path = '/content/drive/MyDrive/new_ligands_csvs_folder/DOYMUC_LFeOOH_CH4_rxt_scan_cont2_TS_stab_ts_irc_reverse_opt_tot_ener.txt'  # Replace 'path_to_data.txt' with the actual file path

# Get original and mirrored coordinates
x, y, x_mirror = mirror_coordinates(file_path)

# Plotting both the original and mirrored coordinates
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-', color='blue', label='Original Data')
plt.plot(x_mirror, y, marker='o', linestyle='-', color='red', label='Mirrored Data')
plt.xlabel('Intrinsic Reaction Coordinate (IRC)')
plt.ylabel('Total Energy (Hartree)')
plt.title('Original Data and its Mirrored Version')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Function to mirror coordinates along X-axis and save the mirrored data to a file
def mirror_coordinates(file_path, output_file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    x = []
    y = []
    for line in lines:
        if not line.startswith('#'):
            coords = line.split()
            x.append(float(coords[0]))
            y.append(float(coords[1]))

    # Create a mirrored version of the coordinates along the X-axis
    x_mirror = [-val for val in x]  # Mirror the x-coordinates

    # Write mirrored coordinates to a new file
    with open(output_file_path, 'w') as output_file:
        for i in range(len(x_mirror)):
            output_file.write(f"{x_mirror[i]}\t{y[i]}\n")

    return x, y, x_mirror

# File paths for the dataset and the output mirrored coordinates file
file_path = '/content/drive/MyDrive/new_ligands_csvs_folder/DOYMUC_LFeOOH_CH4_rxt_scan_cont2_TS_stab_ts_irc_reverse_opt_tot_ener.txt'
output_file_path = '/content/drive/MyDrive/new_ligands_csvs_folder/mirrored_data.txt'  # Replace with your desired file path and name

# Get original and mirrored coordinates and plot them
x, y, x_mirror = mirror_coordinates(file_path, output_file_path)

# Plotting both the original and mirrored coordinates
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-', color='blue', label='Original Data')
plt.plot(x_mirror, y, marker='o', linestyle='-', color='red', label='Mirrored Data')
plt.xlabel('Intrinsic Reaction Coordinate (IRC)')
plt.ylabel('Total Energy (Hartree)')
plt.title('Original Data and its Mirrored Version')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Function to read coordinates from a file and plot
def plot_coordinates_from_file(file_path, title, color='blue'):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    x = []
    y = []
    for line in lines:
        if not line.startswith('#'):
            coords = line.split()
            x.append(float(coords[0]))
            y.append(float(coords[1]))

    plt.plot(x, y, marker='o', linestyle='-', color=color, label=title)

# File paths for the two datasets

file_path_data_1 = '/content/drive/MyDrive/new_ligands_csvs_folder/DOYMUC_LFeOOH_CH4_rxt_scan_cont2_TS_stab_ts_irc_forward_opt_tot_ener.txt'
file_path_data_2 = '/content/drive/MyDrive/new_ligands_csvs_folder/mirrored_data.txt'  # Replace with your desired file path and name

plt.figure(figsize=(8, 6))

# Plotting the first set of coordinates
plot_coordinates_from_file(file_path_data_1, "Total Energy along IRC - DOYMUC_FORWARD")

# Plotting the second set of coordinates
plot_coordinates_from_file(file_path_data_2, "Total Energy along IRC - DOYMUC_REVERSE", color='red')

plt.xlabel('Intrinsic Reaction Coordinate (IRC)')
plt.ylabel('Total Energy (Hartree)')
plt.title('Total Energy along IRC - DOYMUC Complex')
plt.grid(False)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import cv2

# Read the image
image = cv2.imread('content/drive/MyDrive/new_ligands_csvs_folder/DOYMUC_RC.png')  # Replace 'input_image.png' with the path to your image

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to create a mask
_, mask = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY)

# Invert the mask
mask = cv2.bitwise_not(mask)

# Apply the mask to the original image
result = cv2.bitwise_and(image, image, mask=mask)

# Save the resulting image
cv2.imwrite('output_image.png', result)  # Change 'output_image.png' to the desired output file name
```
